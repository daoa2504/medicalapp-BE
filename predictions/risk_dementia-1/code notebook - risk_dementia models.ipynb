In this part we will choose the relevent columns for our model, these columns should have a correlation value higher than 50%

```python
from sklearn.metrics import median_absolute_error, mean_absolute_error, mean_squared_error
import numpy as np
def print_error_metrics(y_true: np.array, y_pred: np.array) -> None:
    """
    Function to compute the mean absolute error, mean squared error, and median absolute error and print to screen
    
    Inputs:
        y_true -> array containing actual dependant variable
        y_pred -> array containing model predictions
    """
    mee = median_absolute_error(y_true,y_pred)
    mae = mean_absolute_error(y_true,y_pred)
    mse = mean_squared_error(y_true,y_pred)
    print("The median absolute error is: {:.2f}".format(mee))
    print("The mean absolute error is: {:.2f}".format(mae))
    print("The root mean squared error is: {:.2f}".format(np.sqrt(mse)))

```

```python
def pred_to_proba(y_pred):
    predictions = []
    for value in y_pred:
        value = int(value*100)
        if value < 0 :
            value = 0
        elif value > 100 :
            value = 100
        else:
            pass
        predictions.append(value/100)
    
    return predictions



```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor 
import catboost as cb

def train_best_reg_models(df, relevant_columns):
    X_train, X_test, y_train, y_test = train_test_split(df[relevant_columns],df["risk_dementia"],test_size=0.8)
    XGB_reg = XGBRegressor().fit(X_train, y_train)
    y_predict = pred_to_proba(XGB_reg.predict(X_test))
    print("Error metrics for XGBoost regression model :")
    print_error_metrics(y_test, y_predict)

    LGBM_reg = LGBMRegressor(metric='rmse')
    LGBM_reg.fit(X_train, y_train)
    y_predict = pred_to_proba(LGBM_reg.predict(X_test))
    print("Error metrics for LightGBM regression model :")
    print_error_metrics(y_test, y_predict)

    model_CB = cb.CatBoostRegressor()
    model_CB.fit(X_train, y_train)
    y_predict = pred_to_proba(model_CB.predict(X_test))
    print("Error metrics for Cat boost regression model :")
    print_error_metrics(y_test, y_predict)
    return XGB_reg, LGBM_reg, model_CB


```

```python
XGB_reg_basic, LGBM_reg_basic, model_CB_basic = train_best_reg_models(df_basic, relevant_columns_basic)
```

```python
XGB_reg_all, LGBM_reg_all, model_CB_all = train_best_reg_models(df_all, relevant_columns_all)
```

```python
XGB_reg_all_plus_plus, LGBM_reg_all_plus_plus, model_CB_all_plus_plus = train_best_reg_models(df_all_plus_plus, relevant_columns_all_plus_plus)
```
